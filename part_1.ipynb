2\. Рассчитайте доли классов `отток` и `не отток`.

```python
import numpy as np
```

```python
churn = np.sum(y.churn == 1) / len(y)
not_churn = np.sum(y.churn == -1) / len(y)
print('Доля классов отток: %f' % churn)
print('Доля классов не отток: %f' % not_churn)
```

3\. Рассчитайте корреляции переменных с целевой функцией и проанализируйте полученные данные.

Обратите внимание на то, что целевая функция предсталвляет собой бинарную переменную. Теоретически, это не помешает рассчитать, например, корреляцию Пирсона между непрерывной переменной и целевой функцией, онднако анализиоовать полученные результаты будет сложно, корреляция Пирсона просто не рассчитана на такое применение. Кто уже забыл, как действовать в такой ситуации - вспоминить можно вот здесь: https://www.coursera.org/teach/stats-for-data-analysis/content/edit/lecture/prX3S/video-subtitles 

```python
corr = pd.DataFrame(columns=['y'], index=X.columns) # Создание датафрейма, в котором будут указаны корреляции переменных
                                                    # с целевой функцией
for var in X.columns[:190]:
    corr.loc[var] = X[y.churn == 1][var].mean() - X[y.churn == -1][var].mean() # Мерой взаимосвязи между числовой и 
                                                                               # бинарной переменными является разность
                                                                               # мат. ожиданий
```

```python
from scipy import stats
```

```python
X_cat = X[X.columns[190:]].fillna('NA').astype(str) # Для категориальных признаков интерпретируем пропущенные значения как
                                                    # отдельная категория
```

```python
for var in X_cat.columns:
    table = pd.crosstab(X_cat[var], y.churn)
    stat, p, dof, expected = stats.chi2_contingency(table)
    corr.loc[var] = np.sqrt(stat / len(X_cat)) # Мерой взаимодействия между двумя категориальными переменными является
                                               # коэффициент V Крамера.
```

```python
corr
```

```python
corr_num = corr[:190]
corr_num[corr_num.y > 0]
```

```python
corr[190:].y.sort_values(ascending=False)
```

Глядя на таблицу корреляций видно, что некоторые признаки имеют все пропущенные значения (например, Var8, Var230).
